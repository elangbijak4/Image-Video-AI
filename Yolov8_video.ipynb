```python
# Langkah 1: Instal library ultralytics
!pip install ultralytics

# Langkah 2: Import library YOLO dari ultralytics dan library lainnya
from ultralytics import YOLO
import cv2
import torch

# Langkah 3: Memuat model YOLOv8 yang telah dilatih
model = YOLO('yolov8n.pt')  # atau 'yolov8s.pt', 'yolov8m.pt', 'yolov8l.pt', 'yolov8x.pt'
```

```python
# Langkah 4: Mengunduh video contoh
!wget -O video.mp4 https://ia800102.us.archive.org/23/items/SampleVideo1280x72020mb/SampleVideo_1280x720_20mb.mp4

# Langkah 5: Membaca video dan melakukan deteksi objek pada setiap frame
cap = cv2.VideoCapture('video.mp4')

# Mendapatkan lebar dan tinggi frame
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Menyiapkan video writer untuk menyimpan hasil
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('output.mp4', fourcc, 20.0, (width, height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Deteksi objek dalam frame
    results = model(frame)

    # Menggambar bounding box pada frame
    annotated_frame = results[0].plot()

    # Menulis frame hasil ke video output
    out.write(annotated_frame)

cap.release()
out.release()

# Langkah 6: Menampilkan video hasil
from IPython.display import Video
Video("output.mp4", embed=True)
```
