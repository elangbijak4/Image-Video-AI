```python
# Langkah 1: Instal library ultralytics
!pip install ultralytics

# Langkah 2: Import library YOLO dari ultralytics
from ultralytics import YOLO
from IPython.display import display, Image  # Untuk menampilkan gambar di Colab

# Langkah 3: Memuat model YOLOv8 yang telah dilatih
model = YOLO('yolov8n.pt')  # atau 'yolov8s.pt', 'yolov8m.pt', 'yolov8l.pt', 'yolov8x.pt'
```

```python
# Langkah 4: Melakukan deteksi pada gambar contoh
# Unduh gambar contoh
!wget -O image.jpg https://ultralytics.com/images/zidane.jpg
img = 'image.jpg'

# Deteksi objek dalam gambar
results = model(img)
```

```python
# Menyimpan hasil deteksi
results_path = results[0].save()

# Menampilkan gambar hasil
display(Image(filename=results_path))
```
